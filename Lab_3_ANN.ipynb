```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow import keras
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/vibnn/Datasets/main/heart.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.drop('target', axis=1)
Y = df['target']

standard_scaler = StandardScaler()
X = standard_scaler.fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
class PrintDot(keras.callbacks.Callback):
    
    def on_epoch_end(self, epoch, logs):
        
        if epoch % 100 == 0: 
            print(' ')
        print('.', end='')
```

```python
model = Sequential()
model.add(Dense(64, input_shape=[x_train.shape[1]]))
model.add(Dense(264, activation='tanh'))
model.add(Dense(124, activation='sigmoid'))
model.add(Dense(64, activation='sigmoid'))
model.add(Dense(1, activation='softmax'))
```

```python
model.compile(optimizer='adam', metrics='binary_accuracy', loss='binary_crossentropy')
```

```python
model.fit(x_train, y_train, epochs=1000, verbose=0, callbacks=[PrintDot()])
```

```python
model.evaluate(x_test, y_test)
```
